```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/AmenaNajeeb/Data/master/bigmart_data.csv")
```

```python
df.head()
```

```python
del df['Item_Identifier']
del df['Outlet_Identifier']
```

```python
df.head()
```

```python
df = df.dropna(how= 'any')
```

```python
df.shape
```

```python
df.head()
```

```python
plt.bar(df['Item_Type'], df['Item_Outlet_Sales'], color = 'purple')
plt.xticks(rotation='vertical')

plt.title("Item Type vs Sales")
plt.ylabel("Sales")
plt.xlabel("Item Type")
```

```python
plt.bar(df['Outlet_Location_Type'], df['Item_Outlet_Sales'])
plt.title("Outlet Location Type vs Sales")
plt.xlabel("Outlet Location Type")
plt.ylabel("Sales")
```

```python
plt.figure(figsize=(15,10))
plt.scatter(df['Item_MRP'], df['Item_Outlet_Sales'], color = 'orange')
plt.title("Item MRP vs Sales")
plt.xlabel("Item MRP")
plt.ylabel("Sales")
```

```python
plt.bar(df['Outlet_Establishment_Year'],df['Item_Outlet_Sales'], color = 'teal')
plt.title("Outlet Est. Year vs Sales")
plt.xlabel("Outlet Est. Year")
plt.ylabel("Sales")
```

```python
plt.bar(df['Outlet_Size'],df['Item_Outlet_Sales'] , color = 'g')
plt.title("Outlet Size vs Sales")
plt.xlabel("Outlet Size")
plt.ylabel("Sales")
```

```python
def label_encode(fat):
  if(fat=='Low Fat' or fat== 'low fat' or fat=='LF'):
    label = 'Low Fat'
  else:
    label = 'Regular'
  return label

df['Item_Fat_Content'] = df['Item_Fat_Content'].apply(label_encode)
```

```python
df.head(10)
```

```python
plt.bar(df['Item_Fat_Content'],df['Item_Outlet_Sales'], color = 'r')
plt.title("Fat Content vs Sales")
plt.xlabel("Fat Content")
plt.ylabel("Sales")
```

```python
plt.figure(figsize=(15,10))
plt.scatter(df['Item_Visibility'],df['Item_Outlet_Sales'], color = 'black')
plt.title("Item Visibility vs Sales")
plt.xlabel("Item Visibility")
plt.ylabel("Sales")
```

```python
plt.bar(df['Outlet_Type'],df['Item_Outlet_Sales'], color = 'brown')
plt.title("Outlet Type vs Sales")
plt.xlabel("Outlet Type")
plt.ylabel("Sales")
```

```python
plt.figure(figsize=(15,10))
plt.scatter(df['Item_Weight'],df['Item_Outlet_Sales'], color = 'grey')
plt.title("Item Weight vs Sales")
plt.xlabel("Item Weight")
plt.ylabel("Sales")
```
